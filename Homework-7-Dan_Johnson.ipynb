```python
import requests

from bs4 import BeautifulSoup
```

```python
#response = requests.get("https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States")
response = requests.get("https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States")
doc = BeautifulSoup(response.text)
```

```python
presidents = doc.find_all('tr')
print(presidents)
```

```python
oval_office=[]

for president in presidents:
    oval_office.append(president.find('b'))
#for oval in oval_office:
    #print(oval)
#print(oval_office)
hail_to_the_chief = []
for oval in oval_office:
    if oval is not None:
        
        big_guy={}
        #print(oval)
        big_guy['name']=oval.find('a', None).get('title',None)
        hail_to_the_chief.append(big_guy)
        
    
print(hail_to_the_chief)
    
    
```
