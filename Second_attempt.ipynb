WANT: titles, subhead, article URL, whether it's premium or not, byline, article type, image URL.

```python
stories=[]
for article in articles:
    temp_story={}
    try:
        temp_story['title'] = article.find(class_="article__title").text
    except:
        print("no title!")
    try:
        temp_story['subhead'] = article.find(class_="article__kicker").text
    except:
        print("There's no subhead!")
    try:
        temp_story['url']=article.get("href")
    except:
        temp_story['url']=article.find("a").get("href")
    try:
        if article.find(class_="sr-only"):
            temp_story['premium'] = True
        else:
            temp_story['premium'] = False
    except:
        print("bing bong!")
    try:
        temp_story['author']=article.find(class_="article__author-name").text
    except:
        print("no author listed!")
    try:
        temp_story['article_type']=article.find(class_="article__type").text
    except:
        print("no article type!")
    try:
        temp_story['img_url']=article.select_one("img[src]").get("src")
    except:
        print("no image url!")
    stories.append(temp_story)
print(stories)        
```
